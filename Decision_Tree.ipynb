```python
#Import library numpy, pandas dan sclikit-learn
import numpy as np
import pandas as pd
from sklearn import tree
```

```python
#Membaca dataset dari file ke pandas dataFrame
irisDataset = pd.read_csv('horses.csv', delimiter=',', header=0)
```

```python
irisDataset.head()
```

```python
#mengubah kelas (kolom "species") dari string ke unique-integer
irisDataset["prize_money"] = pd.factorize(irisDataset.prize_money)[0]
```

```python
irisDataset.head()
```

```python
print(irisDataset)
```

```python
#menghapus kolom "id"
irisDataset = irisDataset.drop(labels="id", axis=1)
```

```python
print(irisDataset)
```

```python
#mengubah dataframe ke array numpy
irisDataset = irisDataset.to_numpy()
```

```python
print(irisDataset)
```

```python
#membagi dataset => 80 baris data untuk training dan 20 baris data untuk testing
dataTraining = np.concatenate((irisDataset[0:20, :], irisDataset[60:100, :]), axis=0)
dataTesting = np.concatenate((irisDataset[20:60, :], irisDataset[100:200, :]), axis=0)                              
```

```python
print(dataTraining)
len(dataTraining)
```

```python
print(dataTesting)
len(dataTesting)
```

```python
#memecah dataset ke input label
inputTraining = dataTraining[:, 0:1]
inputTesting = dataTesting[:, 0:1]
labelTraining = dataTraining[:, 1]
labelTesting = dataTesting[:, 1]
print(labelTraining)
len(labelTraining)
```

```python
#mendefinisikan decision tree classifier
model = tree.DecisionTreeClassifier()
```

```python
#mentraining model
model = model.fit(inputTraining, labelTraining)
```

```python
#memprediksi input data testing
hasilPrediksi = model.predict(inputTesting)
print("Label Sebenarnya : ", labelTesting)
print("Hasil Prediksi : ", hasilPrediksi)
```

```python
#menghitung akurasi
prediksiBenar = (hasilPrediksi == labelTesting).sum()
prediksiSalah = (hasilPrediksi != labelTesting).sum()
print("Prediksi Benar :", prediksiBenar, "data")
print("Prediksi Salah :", prediksiSalah, "data")
print("Akurasi :", prediksiBenar/(prediksiBenar+prediksiSalah) * 100, "%")
```
